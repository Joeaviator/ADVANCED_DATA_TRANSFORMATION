```python
import pandas as pd
import matplotlib.pyplot as plt
import  seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('Titanic Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.drop(columns=['PassengerId', 'Name', 'Ticket'],axis=1)
```

```python
min_age=df['Age'].min()
max_age=df['Age'].max()
print("Min Age: ", min_age)
print("Max Age: ", max_age)
```

```python
bins=[0,20,40,75]
age_labels=["Child", "Adult", "Elderly"]
df['Binned Age']=pd.cut(df['Age'],bins,labels=age_labels)
df.head(10)
```

```python
min_fare=df['Fare'].min()
max_fare=df['Fare'].max()
print("Min Fare: ", min_fare)
print("Max Fare: ", max_fare)
```

```python
bins = [0,50,100,150,275]
fare_labels = ["Low", "Medium", "High", "Very High"]
df['Binned Fare'] = pd.cut(df['Fare'], bins, labels=fare_labels)
df.head(10)
```

```python
#Check the skewedness of age column
sns.distplot(df['Age'])
plt.show()
print(f"Skewness: {df['Age'].skew()}")
```

```python
sns.distplot(df['Fare'])
plt.show()
print(f"Skewness: ",df['Fare'].skew())
```

```python
sns.countplot(df['Binned Age'], palette='rainbow')
plt.show()
```

```python
sns.countplot(df['Binned Fare'], palette='rainbow')
plt.show()
```
